```python
# Define a função para adição
def adicionar(num1, num2):
    return num1 + num2

# Define a função para subtração
def subtrair(num1, num2):
    return num1 - num2

# Define a função para multiplicação
def multiplicar(num1, num2):
    return num1 * num2

# Define a função para divisão, com tratamento para divisão por zero
def dividir(num1, num2):
    if num2 == 0:
        return "Erro! Divisão por zero."
    return num1 / num2

# Imprime as opções de operação para o usuário
print("Selecione a operação:")
print("1. Adicionar")
print("2. Subtrair")
print("3. Multiplicar")
print("4. Dividir")

# Inicia um loop infinito para permitir múltiplas operações
while True:
    # Pede ao usuário para escolher uma operação
    escolha = input("Digite sua escolha (1/2/3/4): ")

    # Verifica se a escolha é uma das opções válidas
    if escolha in ('1', '2', '3', '4'):
        try:
            # Solicita os dois números de entrada ao usuário
            num1 = float(input("Digite o primeiro número: "))
            num2 = float(input("Digite o segundo número: "))
        except ValueError:
            # Lida com entrada inválida (se o usuário não digitar um número)
            print("Entrada inválida. Por favor, digite um número.")
            continue # Volta para o início do loop

        # Executa a operação com base na escolha do usuário e imprime o resultado
        if escolha == '1':
            # Soma e exibe o resultado como inteiro
            print(f"{int(num1)} + {int(num2)} = {int(adicionar(num1, num2))}")

        elif escolha == '2':
            # Subtrai e exibe o resultado como inteiro
            print(f"{int(num1)} - {int(num2)} = {int(subtrair(num1, num2))}")

        elif escolha == '3':
            # Multiplica e exibe o resultado como inteiro
            print(f"{int(num1)} * {int(num2)} = {int(multiplicar(num1, num2))}")

        elif escolha == '4':
            # Divide e exibe o resultado como inteiro, ou a mensagem de erro
            resultado = dividir(num1, num2)
            if isinstance(resultado, str):
                print(resultado)
            else:
                print(f"{int(num1)} / {int(num2)} = {int(resultado)}")

        # Pergunta ao usuário se deseja realizar outra operação
        proxima_operacao = input("Deseja realizar outra operação? (sim/não): ")
        if proxima_operacao.lower() == "não":
            break # Sai do loop se o usuário não quiser continuar
    else:
        # Mensagem para escolha de operação inválida
        print("Escolha inválida. Por favor, digite 1, 2, 3 ou 4.")
```
