```
import os # Importa o módulo os para interagir com o sistema operacional (caminhos, diretórios)
import shutil # Importa o módulo shutil para operações de alto nível em arquivos e diretórios (como mover, copiar)

def organize_files_by_extension(source_directory):
    """
    Organiza arquivos em um diretório, movendo-os para subpastas baseadas em suas extensões.
    Arquivos sem extensão são movidos para uma pasta 'sem_extensao'.
    """
    # Verifica se o diretório de origem fornecido realmente existe
    if not os.path.isdir(source_directory):
        print(f"Erro: O diretório de origem '{source_directory}' não existe.")
        return # Sai da função se o diretório não for válido

    print(f"Organizando arquivos em: {source_directory}")

    # Itera sobre todos os arquivos e pastas dentro do diretório de origem
    for filename in os.listdir(source_directory):
        file_path = os.path.join(source_directory, filename) # Constrói o caminho completo para o item

        # Ignora se o item for um diretório, processando apenas arquivos
        if os.path.isdir(file_path):
            continue

        # Obtém o nome base do arquivo e sua extensão
        name, extension = os.path.splitext(filename)
        extension = extension.lower() # Converte a extensão para minúsculas para padronização (ex: .JPG e .jpg viram .jpg)

        # Decide qual será a pasta de destino com base na extensão
        if not extension:
            # Se não houver extensão (ex: 'LICENSE'), move para a pasta 'sem_extensao'
            destination_folder = os.path.join(source_directory, "sem_extensao")
        else:
            # A pasta de destino será o nome da extensão (sem o ponto inicial)
            destination_folder = os.path.join(source_directory, extension[1:])

        # Cria a pasta de destino se ela não existir. exist_ok=True evita erro se a pasta já existir.
        os.makedirs(destination_folder, exist_ok=True)

        # Move o arquivo para a pasta de destino
        destination_path = os.path.join(destination_folder, filename)
        try:
            shutil.move(file_path, destination_path) # Tenta mover o arquivo
            print(f"Movido: {filename} -> {destination_folder}/") # Informa o sucesso da movimentação
        except Exception as e:
            print(f"Erro ao mover {filename}: {e}") # Imprime mensagem de erro em caso de falha

    print("Organização concluída!")

# --- Exemplo de Uso ---
# As linhas abaixo demonstram como usar a função e criam um ambiente de teste.
# Para testar, execute esta célula completa.

current_dir = os.getcwd() # Obtém o diretório atual onde o notebook está sendo executado
test_dir = os.path.join(current_dir, 'arquivos_para_organizar') # Define o caminho para o diretório de teste

# Limpa o diretório de teste antes de cada execução para garantir um teste limpo
# Isso remove o diretório e todo o seu conteúdo se ele existir
if os.path.exists(test_dir):
    shutil.rmtree(test_dir)

# Cria novamente o diretório de teste. exist_ok=True evita erro se por algum motivo já existir.
os.makedirs(test_dir, exist_ok=True)

# Cria 11 arquivos de teste com diferentes extensões (ou sem extensão) dentro do diretório de teste
with open(os.path.join(test_dir, 'documento.pdf'), 'w') as f: f.write('teste pdf')
with open(os.path.join(test_dir, 'foto.jpg'), 'w') as f: f.write('teste jpg')
with open(os.path.join(test_dir, 'relatorio.docx'), 'w') as f: f.write('teste docx')
with open(os.path.join(test_dir, 'script.py'), 'w') as f: f.write('teste py')
with open(os.path.join(test_dir, 'dados.csv'), 'w') as f: f.write('teste csv')
with open(os.path.join(test_dir, 'apresentacao.pptx'), 'w') as f: f.write('teste pptx')
with open(os.path.join(test_dir, 'musica.mp3'), 'w') as f: f.write('teste mp3')
with open(os.path.join(test_dir, 'video.mp4'), 'w') as f: f.write('teste mp4')
with open(os.path.join(test_dir, 'planilha.xlsx'), 'w') as f: f.write('teste xlsx')
with open(os.path.join(test_dir, 'arquivo_sem_extensao'), 'w') as f: f.write('teste sem extensao') # Arquivo sem extensão
with open(os.path.join(test_dir, 'outro.JPEG'), 'w') as f: f.write('teste JPEG') # Extensão em maiúsculas

print(f"Criados 11 arquivos de teste em: {test_dir}")

# --- Chame a função com o diretório que você deseja organizar ---
# Aqui, a função é chamada para organizar os arquivos que acabaram de ser criados no test_dir.
organize_files_by_extension('./arquivos_para_organizar')
```

```
import requests # Importa a biblioteca requests para fazer requisições HTTP a serviços web

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço a partir de um CEP.
    Retorna uma string formatada com o endereço ou uma mensagem de erro em caso de falha.
    """
    # Remove quaisquer caracteres não numéricos do CEP digitado (ex: -, ., espaço)
    cep = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP resultante tem o tamanho correto (8 dígitos numéricos)
    if len(cep) != 8:
        return "Erro: O CEP deve conter 8 dígitos numéricos."

    # Constrói a URL da API ViaCEP com o CEP fornecido
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # Faz a requisição HTTP GET para a API ViaCEP
        response = requests.get(url)
        # Levanta um erro HTTP para códigos de status ruins (4xx - erro do cliente, 5xx - erro do servidor)
        response.raise_for_status()

        # Converte a resposta JSON da API para um dicionário Python
        dados_cep = response.json()

        # Verifica se a resposta da API inclui a chave 'erro' indicando que o CEP não foi encontrado
        if 'erro' in dados_cep and dados_cep['erro']:
            return f"Erro: CEP '{cep}' não encontrado."

        # Formata as informações do endereço de maneira legível, usando .get() para evitar KeyError
        # e fornecendo 'N/A' como valor padrão para 'complemento' se não existir
        endereco_formatado = (
            f"CEP: {dados_cep.get('cep')}\n" # Endereço completo do CEP
            f"Logradouro: {dados_cep.get('logradouro')}\n" # Nome da rua, avenida, etc.
            f"Complemento: {dados_cep.get('complemento', 'N/A')}\n" # Informações adicionais como 'lado ímpar'
            f"Bairro: {dados_cep.get('bairro')}\n" # Bairro do endereço
            f"Localidade: {dados_cep.get('localidade')}\n" # Cidade
            f"UF: {dados_cep.get('uf')}\n" # Estado (Unidade Federativa)
            f"IBGE: {dados_cep.get('ibge')}" # Código do município no IBGE
        )
        return endereco_formatado

    except requests.exceptions.HTTPError as http_err:
        # Captura e trata erros que ocorrem devido a respostas HTTP de status ruim (ex: 404, 500)
        return f"Erro HTTP: {http_err} - Não foi possível acessar o serviço ViaCEP."
    except requests.exceptions.ConnectionError as conn_err:
        # Captura e trata erros de conexão (ex: sem internet, DNS não encontrado)
        return f"Erro de Conexão: {conn_err} - Verifique sua conexão com a internet."
    except requests.exceptions.Timeout as timeout_err:
        # Captura e trata erros quando a requisição excede o tempo limite definido
        return f"Erro de Timeout: {timeout_err} - A requisição demorou muito para responder."
    except requests.exceptions.RequestException as req_err:
        # Captura e trata quaisquer outros erros gerais relacionados à biblioteca requests
        return f"Erro na Requisição: {req_err} - Ocorreu um erro inesperado ao consultar o CEP."
    except ValueError:
        # Captura erros caso a resposta da API não seja um JSON válido ou esteja vazia
        return "Erro: Resposta inválida da API. Não foi possível decodificar o JSON."
    except Exception as e:
        # Captura e trata qualquer outra exceção inesperada que possa ocorrer
        return f"Erro inesperado: {e}"

# --- Exemplo de Uso ---
# O bloco abaixo é executado apenas quando o script é rodado diretamente (não importado como módulo)
if __name__ == "__main__":
    # Solicita ao usuário que digite um CEP para consulta
    cep_digitado = input("Digite o CEP (apenas números ou com traço): ")

    # Chama a função para consultar o CEP e imprime o resultado formatado
    resultado = consultar_cep(cep_digitado)
    print("\n--- Informações do Endereço ---")
    print(resultado)

    # --- Exemplos de Teste ---
    # Exemplo com um CEP que provavelmente não existe para demonstrar o tratamento de erro de CEP não encontrado
    print("\n--- Teste com CEP inválido ---")
    print(consultar_cep("99999999"))

    # Exemplo com um formato de CEP incorreto (menos de 8 dígitos) para demonstrar a validação inicial
    print("\n--- Teste com CEP em formato incorreto ---")
    print(consultar_cep("123"))

    # Exemplo com um CEP contendo caracteres não numéricos para demonstrar a limpeza do input
    print("\n--- Teste com CEP com caracteres ---")
    print(consultar_cep("01001-000a"))
```

```
import time # Importa o módulo time para funções relacionadas a tempo, como sleep e formatação de data/hora
import random # Importa o módulo random para gerar números aleatórios, usado na simulação de temperatura

# --- Configurações ---
# Define o limite de temperatura em °C. Acima deste valor, um alerta será disparado.
CRITICAL_TEMPERATURE_THRESHOLD = 30
# Define o intervalo de tempo em segundos entre cada verificação de temperatura.
CHECK_INTERVAL_SECONDS = 5

# --- Função para simular a leitura da temperatura ---
def read_temperature():
    """
    Simula a leitura de uma temperatura de um sensor.
    Em um cenário real, esta função seria substituída pela leitura de um hardware real.
    """
    # Gera uma temperatura aleatória entre 20 e 35 graus Celsius, arredondada para duas casas decimais.
    return round(random.uniform(20, 35), 2)

# --- Função para enviar alertas ---
def send_alert(message, alert_type="console"):
    """
    Envia uma mensagem de alerta. Pode ser exibida no console ou simular o envio de um e-mail.

    Args:
        message (str): A mensagem de alerta a ser enviada.
        alert_type (str): O tipo de alerta ('console' para imprimir ou 'email_simulado' para simular e-mail).
    """
    # Obtém o timestamp atual formatado para inclusão na mensagem de alerta.
    timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
    if alert_type == "console":
        print(f"[{timestamp}] ALERTA CRÍTICO: {message}")
    elif alert_type == "email_simulado":
        print(f"[{timestamp}] E-MAIL SIMULADO ENVIADO: Assunto: ALERTA CRÍTICO! Mensagem: {message}")
        # OBS: Em um sistema real, aqui você integraria uma biblioteca ou serviço para envio de e-mails de verdade.
        # Exemplos: 'smtplib' para e-mails via SMTP, ou APIs de serviços como SendGrid, Mailgun, etc.
    else:
        # Caso um tipo de alerta desconhecido seja fornecido.
        print(f"[{timestamp}] Tipo de alerta desconhecido: {alert_type} - {message}")

# --- Loop principal de monitoramento autônomo ---
def autonomous_monitor():
    print("\n--- Iniciando monitoramento autônomo de temperatura ---")
    print(f"Alerta disparado se a temperatura exceder {CRITICAL_TEMPERATURE_THRESHOLD}°C.")
    print(f"Verificação a cada {CHECK_INTERVAL_SECONDS} segundos. Pressione Ctrl+C para parar.")

    # Variável de estado para controlar se um alerta já foi disparado. Isso evita alertas repetidos para a mesma condição.
    last_alert_state = False

    try:
        # Loop infinito para monitoramento contínuo
        while True:
            current_temperature = read_temperature() # Lê a temperatura atual (simulada)

            # Verifica se a temperatura atual excedeu o limite crítico
            if current_temperature > CRITICAL_TEMPERATURE_THRESHOLD:
                # Se a temperatura está crítica e ainda não disparamos um alerta para essa condição
                if not last_alert_state:
                    alert_message = (f"Temperatura atual de {current_temperature}°C "
                                     f"excede o limite crítico de {CRITICAL_TEMPERATURE_THRESHOLD}°C!")
                    send_alert(alert_message, alert_type="console") # Envia o alerta para o console
                    # send_alert(alert_message, alert_type="email_simulado") # Descomente esta linha para testar o envio de e-mail simulado
                    last_alert_state = True # Atualiza o estado para indicar que um alerta foi disparado
                else:
                    # Opcional: Se a condição crítica persiste, loga um aviso sem disparar um novo alerta "crítico"
                    print(f"[{time.strftime("%Y-%m-%M %H:%M:%S")}] "
                          f"AVISO: Condição crítica persistente. Temp: {current_temperature}°C")
            else:
                # Se a temperatura não está crítica
                # E se antes estava crítica (last_alert_state é True), indica que a condição foi resolvida
                if last_alert_state:
                    print(f"[{time.strftime("%Y-%m-%M %H:%M:%S")}] "
                          f"INFO: Temperatura voltou ao normal: {current_temperature}°C.")
                    last_alert_state = False # Reseta o estado do alerta
                else:
                    # Se a temperatura está normal e já estava normal
                    print(f"[{time.strftime("%Y-%m-%M %H:%M:%S")}] "
                          f"INFO: Temperatura atual: {current_temperature}°C. Tudo normal.")

            # Pausa o monitoramento pelo tempo definido em CHECK_INTERVAL_SECONDS
            time.sleep(CHECK_INTERVAL_SECONDS)

    except KeyboardInterrupt:
        # Captura a interrupção pelo teclado (Ctrl+C) para encerrar o monitoramento de forma graciosa.
        print("\n--- Monitoramento interrompido pelo usuário. ---")
    except Exception as e:
        # Captura qualquer outra exceção inesperada que possa ocorrer durante o monitoramento.
        print(f"\n--- ERRO INESPERADO: {e} ---\nMonitoramento encerrado.")

# --- Iniciar o monitoramento ---
# O bloco abaixo garante que a função autonomous_monitor() seja chamada apenas quando o script for executado diretamente.
if __name__ == "__main__":
    autonomous_monitor()
```
