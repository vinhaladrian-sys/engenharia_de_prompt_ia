```
# Script Python para verificar a temperatura de um sistema
try: # Inicia um bloco 'try' para lidar com possíveis erros de entrada
    # Solicita ao usuário que digite a temperatura atual
    # A função input() retorna uma string, então int() tenta convertê-la para um número inteiro
    temperatura = int(input("Digite a temperatura atual: "))

    # Inicia uma condição 'if' para verificar se a temperatura excede 30°C
    if temperatura > 30:
        # Se a condição for verdadeira, imprime um alerta de superaquecimento
        # Usa uma f-string para incluir o valor da variável 'temperatura' na mensagem
        print(f"ALERTA: Temperatura ({temperatura}°C) excede 30°C! Superaquecimento detectado.")
    else:
        # Se a condição 'if' for falsa (temperatura <= 30), executa este bloco
        # Imprime a temperatura atual e informa que está tudo normal
        print(f"Temperatura atual: {temperatura}°C. Tudo normal.")
except ValueError: # Captura o erro específico 'ValueError' se a conversão int() falhar
    # Se o usuário digitar algo que não seja um número inteiro, imprime uma mensagem de erro
    print("Erro: Entrada inválida. Por favor, digite um número para a temperatura.")
```
